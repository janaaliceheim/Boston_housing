### Exercise 3:

```python
# Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)


test=scipy.stats.levene(boston_df[boston_df['CHAS'] == 1.0]['MEDV'],
                   boston_df[boston_df['CHAS'] == 0.0]['MEDV'], center='mean')
print(test)
# since the p-value is smaller than 0.05 we can't assume assume equality of variance (equal variance will be set to False)

ttest=scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 1.0]['MEDV'],
                   boston_df[boston_df['CHAS'] == 0.0]['MEDV'], equal_var = False)

print(ttest)

# Answer: since the pvalue is smaller than 0.05/2 (two sided test) we can conclude that there is a significant difference between Charles river bound houses and those who are not.

```

```python
# Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

# 1 way anova assumed

# two way
youngest = boston_df[boston_df['DISC_AGE'] == '35 years and younger']['MEDV']
medium = boston_df[boston_df['DISC_AGE'] == 'between 35 and 70 years']['MEDV']
oldest = boston_df[boston_df['DISC_AGE'] == '70 years and older']['MEDV']

f_statistic, p_value = scipy.stats.f_oneway(youngest, medium, oldest)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))

# Answer: there is at least one of the means different than the other since the p value is smaller than 0.05
```

```python
# Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

#H0 no correlation!
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])

# Answer: Since the p-value is smaller than 0.05 we can conclude that there IS a correlation between the two variables.
```

```python
# What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

## X is the input variables (or independent variables)
X = boston_df['DIS']
## y is the target/dependent variable
y = boston_df['MEDV']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()


# Answer: The impact of an additional weighted distance is 1.0916 per added weighted distance.
```
